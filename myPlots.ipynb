```python
def myPlot(FC, row_list, col_list, colorbar_type='all'):
    
    import numpy as np
    import matplotlib.pyplot as plt
    
    font = {'family' : 'serif',
    'size'  : 10}
    plt.rc('font', **font)
    titlefontdict = {'fontsize': 10}


    n_row = len(row_list)
    n_col = len(col_list)
    n_col_mod = n_col
    for i in range(n_col):
        if len(col_list[i]) == 0:
            n_col_mod = n_col_mod - 1
            
    
    # ============= Designating each FC to an Image =============
    imag = []
    Diff = []
    for i in range(n_row):
        FC_measure, SNR,  seq = row_list[i]
        for j in range(n_col):
            Data_type = col_list[j]
            if len(Data_type) == 0: continue
            
            if seq == 'task-rest':
                if Data_type == 'physio-raw':
                    imag = imag + [(FC[FC_measure][SNR]['physio']['task'] - FC[FC_measure][SNR]['physio']['rest']) - (FC[FC_measure][SNR]['raw']['task'] - FC[FC_measure][SNR]['raw']['rest'])]
                    # Diff_basket = np.linalg.norm( imag[-1] )
                    Diff_basket = np.linalg.norm( imag[-1] ) / np.linalg.norm( (FC[FC_measure][SNR]['raw']['task'] - FC[FC_measure][SNR]['raw']['rest']) )
                    Diff_basket = np.linalg.norm( imag[-1] ) / np.linalg.norm( (FC[FC_measure][SNR]['raw']['task'] - FC[FC_measure][SNR]['raw']['rest']) )
                    Diff = Diff + [Diff_basket]
                elif Data_type == 'physio_task_mdul-raw':
                    imag = imag + [(FC[FC_measure][SNR]['physio_task_mdul']['task'] - FC[FC_measure][SNR]['physio_task_mdul']['rest']) - (FC[FC_measure][SNR]['raw']['task'] - FC[FC_measure][SNR]['raw']['rest'])]
                    # Diff_basket = np.linalg.norm( imag[-1] )
                    Diff_basket = np.linalg.norm( imag[-1] ) / np.linalg.norm( (FC[FC_measure][SNR]['raw']['task'] - FC[FC_measure][SNR]['raw']['rest']) )
                    Diff = Diff + [Diff_basket]
                elif Data_type == 'physio_task_mdul-physio':
                    imag = imag + [(FC[FC_measure][SNR]['physio_task_mdul']['task'] - FC[FC_measure][SNR]['physio_task_mdul']['rest']) - (FC[FC_measure][SNR]['physio']['task'] - FC[FC_measure][SNR]['physio']['rest'])]
                    # Diff_basket = np.linalg.norm( imag[-1] )
                    Diff_basket = np.linalg.norm( imag[-1] ) / np.linalg.norm( (FC[FC_measure][SNR]['physio']['task'] - FC[FC_measure][SNR]['physio']['rest']) )
                    Diff = Diff + [Diff_basket]
                else:
                    imag = imag + [(FC[FC_measure][SNR][Data_type]['task'] - FC[FC_measure][SNR][Data_type]['rest'])]
                    # Diff_basket = np.linalg.norm( imag[-1] - (FC[FC_measure][SNR]['raw']['task'] - FC[FC_measure][SNR]['raw']['rest']) )
                    Diff_basket = np.linalg.norm( imag[-1] - (FC[FC_measure][SNR]['raw']['task'] - FC[FC_measure][SNR]['raw']['rest']) ) / np.linalg.norm( (FC[FC_measure][SNR]['raw']['task'] - FC[FC_measure][SNR]['raw']['rest']) )
                    Diff = Diff + [Diff_basket]
            else:
                if Data_type == 'physio-raw':
                    imag = imag + [(FC[FC_measure][SNR]['physio'][seq] - FC[FC_measure][SNR]['raw'][seq])]
                    # Diff_basket = np.linalg.norm( imag[-1] )
                    Diff_basket = np.linalg.norm( imag[-1] ) / np.linalg.norm( FC[FC_measure][SNR]['raw'][seq] )
                    Diff = Diff + [Diff_basket]
                elif Data_type == 'physio_task_mdul-raw':
                    imag = imag + [(FC[FC_measure][SNR]['physio_task_mdul'][seq] - FC[FC_measure][SNR]['raw'][seq])]
                    # Diff_basket = np.linalg.norm( imag[-1] )
                    Diff_basket = np.linalg.norm( imag[-1] ) / np.linalg.norm( FC[FC_measure][SNR]['raw'][seq] )
                    Diff = Diff + [Diff_basket]
                elif Data_type == 'physio_task_mdul-physio':
                    imag = imag + [(FC[FC_measure][SNR]['physio_task_mdul'][seq] - FC[FC_measure][SNR]['physio'][seq])]
                    # Diff_basket = np.linalg.norm( imag[-1] )
                    Diff_basket = np.linalg.norm( imag[-1] ) / np.linalg.norm( FC[FC_measure][SNR]['physio'][seq] )
                    Diff = Diff + [Diff_basket]
                else:
                    imag = imag + [(FC[FC_measure][SNR][Data_type][seq])]
                    # Diff_basket = np.linalg.norm( imag[-1] - FC[FC_measure][SNR]['noisy'][seq] )
                    Diff_basket = np.linalg.norm( imag[-1] - FC[FC_measure][SNR]['raw'][seq] ) / np.linalg.norm( FC[FC_measure][SNR]['raw'][seq] )
                    Diff = Diff + [Diff_basket]

    # ============= Determining colorbar's limits =============
    n_fig = len(imag)
    if colorbar_type == 'all':
        colorbar_limits = 0
        for i in range(n_fig):
            colorbar_limits = np.max( [colorbar_limits, abs( imag[i] - np.diag(np.diag(imag[i])) ).max()] )
    elif colorbar_type == 'row':
        colorbar_limits = np.zeros(n_row)
        for i in range(n_row):
            for j in range(n_col_mod):
                if len(col_list[j]) == 0: continue
                colorbar_limits[i] = np.max( [colorbar_limits[i], abs( imag[j+i*n_col_mod] - np.diag(np.diag(imag[j+i*n_col_mod])) ).max() ] )
    else: print(f"'colorbar_type' should be either 'all' or 'row'!")
        
    # ============= Ploting =============
    k = 0
    l = 0
    for i in range(n_row):
        FC_measure, SNR,  seq = row_list[i]
        for j in range(n_col):
            Data_type = col_list[j]
            if len(Data_type) == 0:
                l = l+1
                continue
            
            plt.subplot(n_row, n_col, k+1+l )
            if colorbar_type == 'all':
                plt.imshow(imag[k] - np.diag( np.diag( imag[k] ) ), cmap='seismic', vmax=colorbar_limits, vmin=-colorbar_limits)
            elif colorbar_type == 'row':
                plt.imshow(imag[k] - np.diag( np.diag( imag[k] ) ), cmap='seismic', vmax=colorbar_limits[i], vmin=-colorbar_limits[i])
            plt.title(f"(rError={Diff[k].round(2)})", titlefontdict)
            plt.title(f"FC: {Data_type}; (rError={Diff[k].round(2)})", titlefontdict)
            if (k+l+1)%n_col == 1: plt.ylabel(f"{FC_measure}; SNR={SNR}; {seq}")
            
            plt.hlines(np.arange(m)-.5, -np.ones(m)*.5, np.ones(m)*(m-.5), linestyles='solid', colors='k')
            plt.vlines(np.arange(m)-.5, -np.ones(m)*.5, np.ones(m)*(m-.5), linestyles='solid', colors='k')
            plt.yticks(np.arange(m), np.arange(1, m+1))
            plt.xticks(np.arange(m), np.arange(1, m+1))
            plt.colorbar()
            k = k+1
```

```python
def myScatter(x, y, plt_title, x_label, y_label, data_label, data_color='blue'):
    import numpy as np
    import statsmodels.api as sm
    import matplotlib.pyplot as plt
    
    Y = y.flatten()
    X = x.flatten()
    X2 = sm.add_constant(X)
    est = sm.OLS(Y, X2).fit().summary2()
    r2 = float(est.tables[0][1][6])
    est = est.tables[1]
    b = [est['Coef.']['const'], est['Coef.']['x1']]
    p_val = np.array([est['P>|t|']['x1'], est['P>|t|']['const']]).round(3)
    y_est = b[0] + b[1] * np.linspace(x.min(), x.max(), 100)
    
    lbl = f"{b[1].round(2)}x + {b[0].round(2)} \n (pVal = {p_val})\n (r2 = {r2})"

    plt.scatter(x, y, color=data_color, label=data_label)
    plt.plot(np.linspace(x.min(), x.max(), 100), y_est, linestyle='--', color='black', label=lbl)
    plt.title(plt_title)
    plt.ylabel(y_label)
    plt.xlabel(x_label)
    plt.legend()
    plt.grid()
```

```python
def graph_show(A, edge_width=False, edge_label=False):
    import numpy as np
    import matplotlib.pyplot as plt
    import networkx as nx
    _, m = A.shape

    G = nx.DiGraph()

    node_list = np.arange(1,m+1)
    for i in node_list:
        G.add_node(i)
    
    pos=nx.circular_layout(G)
    # pos=nx.kamada_kawai_layout(G)

    edge_weights = {}
    for i in node_list:
        for j in node_list:
            if A[i-1, j-1]==0: continue
            G.add_edge(j, i, weight=A[i-1, j-1])
            edge_weights[i,j] = A[i-1, j-1].round(2)

    nx.draw_networkx_nodes(G,pos,node_color='tab:blue')
    nx.draw_networkx_labels(G,pos, font_family='serif')

    if edge_width==False:
        for (node1, node2, weight) in G.edges(data=True):
            nx.draw_networkx_edges(G,pos, [(node1, node2)])
    if edge_width==True:
        for (node1, node2, weight) in G.edges(data=True):
            nx.draw_networkx_edges(G,pos, [(node1, node2)], width=weight['weight']/max(list(edge_weights.values()))*2)
    if edge_label==True: 
        nx.draw_networkx_edge_labels(G,pos, edge_labels=edge_weights, label_pos=0.5)

    plt.axis('off')
```

```python
def graph_show_input(A, B, edge_width=False, edge_label=False):
    import numpy as np
    import matplotlib.pyplot as plt
    import networkx as nx
    _, m_a = A.shape
    m_b, _ = B.shape
    m = m_a + m_b

    AB = np.zeros([m, m])
    AB[:m_a, :m_a] = A
    AB[:m_a, m_a:] = B.T

    G = nx.DiGraph()

    node_list = np.arange(1,m+1)
    for i in node_list:
        G.add_node(i)

    pos=nx.circular_layout(G)

    edge_weights = {}
    for i in node_list:
        for j in node_list:
            if AB[i-1, j-1]==0: continue
            G.add_edge(j, i, weight=AB[i-1, j-1])
            edge_weights[i,j] = AB[i-1, j-1].round(2)

    node_label = {}
    for i in range(1,m_a+1):
        node_label[i] = f"{i}"
    for i in range(m_a+1, m+1):
        node_label[i] = f"u{i-m_a}"
        

    nx.draw_networkx_nodes( G, pos, nodelist=node_list[ :m_a], node_color ='tab:blue', node_shape='o')
    nx.draw_networkx_nodes( G, pos, nodelist=node_list[m_a:m], node_color ='tab:blue', node_shape='s')
    nx.draw_networkx_labels(G, pos, labels=node_label, font_family='serif')


    if edge_width==False:
        for (node1, node2, weight) in G.edges(data=True):
            nx.draw_networkx_edges(G,pos, [(node1, node2)])
    if edge_width==True:
        for (node1, node2, weight) in G.edges(data=True):
            nx.draw_networkx_edges(G,pos, [(node1, node2)], width=weight['weight']/max(list(edge_weights.values()))*2)
    if edge_label==True: 
        nx.draw_networkx_edge_labels(G,pos, edge_labels=edge_weights, label_pos=0.5)

    plt.axis('off')
```
